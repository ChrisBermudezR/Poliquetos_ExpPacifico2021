```python
import pandas as pd
from scipy import stats

df = pd.read_csv('https://gitlab.com/ChrisBermudezR/metadinophysis/-/raw/main/Analisis/datosAbundancia.csv')

df


Malaga = df[df['Bahia']=='Malaga']
Buenaventura = df[df['Bahia']=="Buenaventura"]

Baja = df[df['Marea']=='Baja']
Alta = df[df['Marea']=="Alta"]

print(stats.mannwhitneyu(Malaga['Abundancia'], Buenaventura['Abundancia'], use_continuity=True, alternative='two-sided', axis=0, method='auto'))
print(stats.mannwhitneyu(Baja['Abundancia'], Alta['Abundancia'], use_continuity=True, alternative='two-sided', axis=0, method='auto'))

```
